```python
def get_lexicon_model(lexicon_dict):
    ANGER = 0
    ANTICIPATION = 0
    DISGUST = 0
    FEAR = 0
    JOY = 0
    SADNESS = 0
    SURPRISE = 0
    TRUST = 0

    new_dict = dict()

    for index, val in enumerate(lexicon_dict):
    #     print(val)
        print(val + ":" +  str(lexicon_dict[val]))

        ANGER += lexicon_dict[val]['anger']
        ANTICIPATION += lexicon_dict[val]['anticipation']
        DISGUST += lexicon_dict[val]['disgust']
        FEAR += lexicon_dict[val]['fear']
        JOY += lexicon_dict[val]['joy']
        SADNESS += lexicon_dict[val]['sadness']
        SURPRISE += lexicon_dict[val]['surprise']
        TRUST += lexicon_dict[val]['trust']

    new_dict['anger'] = ANGER
    new_dict['anticipation'] = ANTICIPATION
    new_dict['disgust'] = DISGUST
    new_dict['fear'] = FEAR
    new_dict['joy'] = JOY
    new_dict['sadness'] = SADNESS
    new_dict['surprise'] = SURPRISE
    new_dict['trust'] = TRUST

    return new_dict
```
